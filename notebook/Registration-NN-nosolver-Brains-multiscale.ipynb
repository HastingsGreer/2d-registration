```python
BATCH_SIZE = 4
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
class DiffUNet(nn.Module):
    def __init__(self):
        super(DiffUNet, self).__init__()
        self.num_layers = 5   
        down_channels = [2, 16, 16, 32, 64, 128]
        up_channels_out = [8, 8, 16, 32, 64]
        up_channels_in = [24, 32, 64, 128, 128]
        diff_channels_in = [10, 24, 32, 64, 128, 128]
        
        self.downConvs = nn.ModuleList([])
        self.upConvs = nn.ModuleList([])
        self.diffConvs = nn.ModuleList([])
        for depth in range(self.num_layers):
            self.downConvs.append( 
                nn.Conv3d(
                    down_channels[depth], 
                    down_channels[depth + 1],
                    kernel_size=3,
                    padding=1,
                    stride=2
                )
            )
            self.upConvs.append(
                nn.ConvTranspose3d(
                    up_channels_in[depth], 
                    up_channels_out[depth],
                    kernel_size=4,
                    padding=1,
                    stride=2
                )
            )
            self.diffConvs.append(nn.Conv3d(
                diff_channels_in[depth],
                3,
                kernel_size=3,
                padding=1,
                )
            )
            
        self.firstDiffConv = nn.Conv3d(
            128, 3, kernel_size=3, padding=1)
     
    def forward(self, x, y):
        x = torch.cat([x, y], 1)
        skips = []
        out = []
        for depth in range(self.num_layers):
            skips.append(x)
            x = F.relu(self.downConvs[depth](x))
        diff = self.firstDiffConv(x)
        for depth in reversed(range(self.num_layers)):
            x = F.relu(self.upConvs[depth](x))
            x = torch.cat([x, skips[depth]], 1)
            diff = (
                F.interpolate(diff, scale_factor=2, mode="trilinear") 
              + self.diffConvs[depth](x) / 5
            )
            out.append(diff)
        return out 

class MermaidNet(nn.Module):
    def __init__(self, image_size):
        super(MermaidNet, self).__init__()       
        assert(len(image_size) == 3)
    
        self.warped_fixed_images = [None for _ in range(4)]
        self.downsampled_moving_images = [None for _ in range(4)]
 
        self.spacings = [None for _ in range(4)]
        self.identity_maps = [None for _ in range(4)]
        for depth in range(4):
            stride = 2**depth
            self.sz = np.array(
                [BATCH_SIZE, 1] + [ i // stride for i in image_size])
            self.spacings[depth] = 1./(self.sz[2::]-1) 

            _id = py_utils.identity_map_multiN(self.sz, self.spacings[depth])
            self.identity_maps[depth] = torch.from_numpy(_id).cuda()
         
        self.regis_net = DiffUNet().cuda()
        
    def forward(self, moving_image, fixed_image):
        self.phis = self.regis_net(moving_image, fixed_image)
        
        self.phiinvs = self.regis_net(fixed_image, moving_image)
        self.all_loss = 0
        
        downsampled_moving_image = moving_image
        downsampled_fixed_image = fixed_image
        
        for depth in [0, 1, 2, 3]:
            stride = 2**depth
            self.phi1 = self.phis[-(1 + depth)] + self.identity_maps[depth]

            self.phiinv = self.phiinvs[-(1 + depth)] + self.identity_maps[depth]
            
            self.downsampled_moving_images[depth] = downsampled_moving_image
            
            self.warped_moving_image = py_utils.compute_warped_image_multiNC(
                downsampled_moving_image, 
                self.phi1, 
                self.spacings[depth], 1)
            
            

            self.warped_fixed_image = py_utils.compute_warped_image_multiNC(
                downsampled_fixed_image, self.phiinv, self.spacings[depth], 1)
            
            self.warped_fixed_images[depth] = self.warped_fixed_image

            self.fake_ident = py_utils.compute_warped_image_multiNC(
                self.phiinv, self.phi1, self.spacings[depth], 1)
            ll = 360
            regularization_loss = ll * torch.mean(
                (self.identity_maps[depth] - self.fake_ident)**2)
            self.distortion_magnitude = ll * torch.mean(
                (self.identity_maps[depth] - self.phi1)**2
            )
            similarity_loss = (
                torch.mean((self.warped_moving_image - downsampled_fixed_image)**2) + 
                torch.mean((self.warped_fixed_image - downsampled_moving_image)**2)
            )
            print("step")
            print("Depth:", depth)
            print(regularization_loss.detach().cpu().item(), similarity_loss.detach().cpu().item(), 
             self.distortion_magnitude.detach().cpu().item())
        
            self.all_loss += regularization_loss + similarity_loss
            
            downsampled_moving_image = F.avg_pool3d(downsampled_moving_image, 2, 2)
            downsampled_fixed_image = F.avg_pool3d(downsampled_fixed_image, 2, 2)

net = MermaidNet([64, 128, 128])
```

```python
optimizer = torch.optim.Adam(net.parameters(), lr=.00005)
```

```python
net.train()
```

```python
def make_batch():
    image = np.array([random.choice(brains)[::2, ::2, ::2] for _ in range(BATCH_SIZE)])
    image = image.reshape(BATCH_SIZE, 1, 64, 128, 128)
    image = torch.tensor(image).cuda()
    return image
for _ in range(10):

    moving_image = make_batch()
    fixed_image = make_batch()
    optimizer.zero_grad()
    net(moving_image, fixed_image)
    loss = net.all_loss


    loss.backward()
    optimizer.step()
```
